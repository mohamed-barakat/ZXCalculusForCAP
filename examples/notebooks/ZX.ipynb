```julia
using CapAndHomalg
```

```julia
LoadPackage( "FunctorCategories" )
```

```julia
decorating_quiver = CreateQuiver( 4, [ 0,1, 0,2, 0,3 ] )
```

```julia
DecoratedFinQuivers =
  CategoryOfDecoratedQuivers( decorating_quiver,
    ## decoration_of_vertices
    [ "black",    # 0
      "red",      # 1
      "green",    # 2
      "orange" ], # 3
    ## decoration_of_arrows
    [ "black",
      "black",
      "black" ]  
    )
```

```julia
B = TerminalObject( DecoratedFinQuivers )
```

```julia
Subobject( B, [ 2 ] )
```

```julia
zx1 = CreateDecoratedQuiver( DecoratedFinQuivers,
               [ 0,0,0,0,0,0,0, 1,1, 2, 3 ], # decoration of [ 0 .. 10 ]
               [ 0,7,0, 1,7,0, 2,7,0, 2,9,1, 3,10,2, 4,10,2, 4,9,1, 5,9,1, 5,8,0, 6,8,0 ] ## arrows_with_decoration
    )
```

```julia
IsWellDefined( zx1 )
```

```julia
i1 = Subobject( zx1, [ 0, 3 ], [ ] )
```

```julia
o1 = Subobject( zx1, [ 1, 6 ], [ ] )
```

```julia
Pushout( o1, i1 )
```

```julia
e1 = InjectionOfCofactorOfPushout( [ o1, i1 ], 1 ); SetIsMonomorphism( e1, true ); e1
```

```julia
e2 = InjectionOfCofactorOfPushout( [ o1, i1 ], 2 ); SetIsMonomorphism( e2, true ); e2
```

```julia
PreCompose( i1, e1 )
```

```julia
PreCompose( o1, e2 )
```
