```julia
H = H_Gate( zx )
```

```julia
X_1_2 = X_Spider( zx, 1, 2 )
```

```julia
Z_2_1 = Z_Spider( zx, 2, 1 )
```

```julia
succ_mod_4 = ( X_Spider( zx, g"π", 1, 1 ) + X_1_2 ) * ( Z_2_1 + id )
```
